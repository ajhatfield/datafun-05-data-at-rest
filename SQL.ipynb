## Section 1: SELECT. Complete 17.2.2 SELECT (1 query)

```python
# Retrieve only columns first and last from authors table

pd.read_sql('SELECT first, last FROM authors', connection)
```

## Section 2: WHERE. Complete 17.2.3 WHERE (3 queries)

```python
# Slect the title, edition, and copyright for all books with copyright years greater than 2016.

pd.read_sql("""SELECT title, edition, copyright FROM titles WHERE copyright > '2016'""", connection)
```

```python
# Locate all the authors whose last name start with the letter D.

pd.read_sql("""SELECT id, first, last FROM authors WHERE last LIKE 'D%'""", connection, index_col = ['id'])
```

```python
# Select the rows of all the authors whose last names start wiht any chartacter, followed by the letter b, 
#followed by any additional characters (specified by %)

pd.read_sql("""SELECT id, first, last FROM authors WHERE first LIKE '_b%'""", connection, index_col=['id'])
```

## Section 3: ORDER BY. Complete 17.2.4 ORDER BY (4 queries)

```python
# Sort the titles in ascending order

pd.read_sql("SELECT title FROM titles ORDER BY title ASC", connection)
```

```python
# sort all authors' names by last name then by first name for any authors who have the same last name. 

pd.read_sql("""SELECT id, first, last FROM authors ORDER BY last, first""", connection, index_col=['id'])
```

```python
# sort the authors in descending order by last name and ascending order by first name for any authors who have the same last name. 

pd.read_sql("""SELECT id, first, last FROM authors ORDER BY last DESC, first ASC""", connection, index_col=['id'])
```

```python
#get the isbn, title, eidition, copyright of each book in the titles table that has a title ending with 'How to Program' and sort
#them in ascending order by title. 

pd.read_sql("""SELECT isbn, title, edition, copyright 
            FROM titles 
            WHERE title LIKE '%How to Program' 
            ORDER BY title""", connection)
```

## Section 4: INNER JOIN. Complete 17.2.5 INNER JOIN (1 query)

```python
pd.read_sql("""SELECT first, last, isbn
            FROM authors
            INNER JOIN author_ISBN
                ON authors.id = author_ISBN.id
            ORDER BY last, first""", connection).head()
```
